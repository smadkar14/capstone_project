```python
####################################### Hindustantimes ##########################################################
```

```python
res4 = requests.get('https://www.hindustantimes.com/')
```

```python
soup4 = BeautifulSoup(res4.text, 'lxml')
```

```python
news_box4=soup4.find('ul',{'class':'latestnews-topblk'})
```

```python
all_news4=news_box4.find_all('a')
```

```python
t=[]
for news4 in all_news4:
    t.append(news4.text)
```

```python
l=[]
for a in news_box4.find_all('a', href=True):
    l.append(a['href'])
```

```python
l
```

```python
a=[]
b=[]
c=[]
for i in l:
    url=i
    article=Article(url)
    article.download()
    article.parse()
    nltk.download('punkt')
    article.nlp()
    a.append(article.text)
    b.append(article.title)
    c.append('Hindustan Times')
```
